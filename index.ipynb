```r
moda = function(x) {
  t = table(x)
  return(as.numeric(names(t)[t == max(t)])) 
}
```

```r
modaclass = function(li, h, F1, F2, F3) {
  modaclass = li + h*(F2 - F1)/((F2 - F1) + F2 - F3) 
  modaclass
}
```

```r
variancia = function(n, m, x) {
  soma = 0
  for (i in 1:n){
    soma = soma + (x[i]-m)^2
  }
  soma
  variancia  = soma/(n-1)
  variancia
}

y = na.omit(DiscoveryYear)
variancia(n, m, y)
```
